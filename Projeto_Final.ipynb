```python
# Conexão
conn = sqlite3.connect("Banco")
cursor = conn.cursor()

# Lista oficial de circunstâncias do óbito (CIRCOBITO)
circunstancias = [
    (1, "Acidente"),
    (2, "Suicídio"),
    (3, "Homicídio"),
    (4, "Outros"),
    (9, "Ignorado"),
]

for cd_circ, descricao in circunstancias:
    cursor.execute('''
        SELECT SK_CIRCUSTANCIA FROM DWCD_CIRCUNSTANCIA
        WHERE CD_CIRCUNSTANCIA = ?
    ''', (cd_circ,))
    existing = cursor.fetchone()

    if not existing:  # Só insere se não existir
        cursor.execute('''
            INSERT INTO DWCD_CIRCUNSTANCIA
            (CD_CIRCUNSTANCIA, DS_CIRCUNSTANCIA, ST_SITUACAO, DT_CARGA)
            VALUES (?, ?, ?, ?)
        ''', (cd_circ, descricao, 'A', datetime.now().strftime("%Y-%m-%d %H:%M:%S")))
        print(f"✅ Inserido: {cd_circ} - {descricao}")
    else:
        print(f"⚠️ Já existe no banco: {cd_circ} - {descricao}")

# Commit e fechar
conn.commit()
conn.close()
```

```python
def merge_cid(df_cid, db_path="Banco"):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Pega dados atuais do banco
    cursor.execute("SELECT CD_CID, DS_DOENÇA, ST_SITUACAO FROM DWCD_DOENÇA")
    db_rows = cursor.fetchall()
    df_db = pd.DataFrame(db_rows, columns=["CD_CID", "DS_DOENÇA", "ST_SITUACAO"])

    # Garantir que código esteja em string sem pontos
    df_cid["codigo"] = df_cid["codigo"].astype(str).str.replace(".", "", regex=False)

    # Data atual
    now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    # --- 1) Inserir novos (estão na origem mas não no banco)
    novos = df_cid[~df_cid["codigo"].isin(df_db["CD_CID"])]
    for _, row in novos.iterrows():
        cursor.execute("""
            INSERT INTO DWCD_DOENÇA (CD_CID, DS_DOENÇA, ST_SITUACAO, DT_CARGA)
            VALUES (?, ?, 'A', ?)
        """, (row["codigo"], row["nome"], now))
        print(f"✅ Inserido novo CID {row['codigo']}")

    # --- 2) Inativar (estão no banco mas não na origem)
    inativos = df_db[~df_db["CD_CID"].isin(df_cid["codigo"])]
    for _, row in inativos.iterrows():
        cursor.execute("""
            UPDATE DWCD_DOENÇA
            SET ST_SITUACAO = 'I', DT_CARGA = ?
            WHERE CD_CID = ?
        """, (now, row["CD_CID"]))
        print(f"⚠️ Inativado CID {row['CD_CID']}")

    # --- 3) Atualizar descrição ou reativar se necessário
    atualizacoes = df_cid.merge(df_db, left_on="codigo", right_on="CD_CID")
    for _, row in atualizacoes.iterrows():
        # Caso a descrição tenha mudado ou esteja inativa mas voltou ao df
        if (row["nome"] != row["DS_DOENÇA"]) or (row["ST_SITUACAO"] == "I"):
            cursor.execute("""
                UPDATE DWCD_DOENÇA
                SET DS_DOENÇA = ?, ST_SITUACAO = 'A', DT_CARGA = ?
                WHERE CD_CID = ?
            """, (row["nome"], now, row["codigo"]))

            if row["ST_SITUACAO"] == "I":
                print(f"♻️ Reativado CID {row['codigo']}")
            else:
                print(f"🔄 Atualizado CID {row['codigo']}")

    conn.commit()
    conn.close()
    print("✅ Merge finalizado com sucesso!")
```

```python
def merge_municipios(df_mun, db_path="Banco"):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Pega dados atuais do banco
    cursor.execute("SELECT CD_MUNICIPIO, NM_MUNICIPIO, DS_PORTE, SK_ESTADO, ST_SITUACAO FROM DWCD_MUNICIPIO")
    db_rows = cursor.fetchall()
    df_db = pd.DataFrame(db_rows, columns=["CD_MUNICIPIO", "NM_MUNICIPIO", "DS_PORTE", "SK_ESTADO", "ST_SITUACAO"])

    # Data atual
    now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    # --- 1) Inserir novos (ignorando -1)
    novos = df_mun[(df_mun["IBGE"] != -1) & (~df_mun["IBGE"].isin(df_db["CD_MUNICIPIO"]))]
    for _, row in novos.iterrows():
        cd_mun = int(row["IBGE"])
        nm_mun = row["Município"]
        porte = row["Porte"]

        # Descobre SK_ESTADO pelo código IBGE
        cd_estado = int(str(row["IBGE"]).zfill(7)[:2])
        cursor.execute("SELECT SK_ESTADO FROM DWCD_ESTADO WHERE CD_ESTADO = ?", (cd_estado,))
        res = cursor.fetchone()
        sk_estado = res[0] if res else -1

        cursor.execute("""
            INSERT INTO DWCD_MUNICIPIO (CD_MUNICIPIO, NM_MUNICIPIO, SK_ESTADO, DS_PORTE, ST_SITUACAO, DT_CARGA)
            VALUES (?, ?, ?, ?, 'A', ?)
        """, (cd_mun, nm_mun, sk_estado, porte, now))
        print(f"✅ Inserido município {cd_mun} - {nm_mun}")

    # --- 2) Inativar (existem no banco mas não no df_mun, ignorando -1)
    inativos = df_db[(df_db["CD_MUNICIPIO"] != -1) & (~df_db["CD_MUNICIPIO"].isin(df_mun["IBGE"]))]
    for _, row in inativos.iterrows():
        cursor.execute("""
            UPDATE DWCD_MUNICIPIO
            SET ST_SITUACAO = 'I', DT_CARGA = ?
            WHERE CD_MUNICIPIO = ?
        """, (now, row["CD_MUNICIPIO"]))
        print(f"⚠️ Inativado município {row['CD_MUNICIPIO']}")

    # --- 3) Atualizar descrição/porte/estado ou reativar (ignorando -1)
    atualizacoes = df_mun.merge(df_db, left_on="IBGE", right_on="CD_MUNICIPIO")
    for _, row in atualizacoes.iterrows():
        if row["IBGE"] == -1:
            continue  # ignora "Não informado"

        cd_mun = int(row["IBGE"])
        nm_mun = row["Município"]
        porte = row["Porte"]

        # Descobre SK_ESTADO atualizado
        cd_estado = int(str(row["IBGE"]).zfill(7)[:2])
        cursor.execute("SELECT SK_ESTADO FROM DWCD_ESTADO WHERE CD_ESTADO = ?", (cd_estado,))
        res = cursor.fetchone()
        sk_estado = res[0] if res else -1

        mudou_nome = (nm_mun != row["NM_MUNICIPIO"])
        mudou_porte = (porte != row["DS_PORTE"])
        mudou_estado = (sk_estado != row["SK_ESTADO"])
        estava_inativo = (row["ST_SITUACAO"] == "I")

        if mudou_nome or mudou_porte or mudou_estado or estava_inativo:
            cursor.execute("""
                UPDATE DWCD_MUNICIPIO
                SET NM_MUNICIPIO = ?, DS_PORTE = ?, SK_ESTADO = ?, ST_SITUACAO = 'A', DT_CARGA = ?
                WHERE CD_MUNICIPIO = ?
            """, (nm_mun, porte, sk_estado, now, cd_mun))

            if estava_inativo:
                print(f"♻️ Reativado município {cd_mun} - {nm_mun}")
            else:
                print(f"🔄 Atualizado município {cd_mun} - {nm_mun}")

    conn.commit()
    conn.close()
    print("✅ Merge de municípios finalizado com sucesso!")
```

```python
def merge_ocupacao(df_cbo, db_path="Banco"):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Pega dados atuais do banco
    cursor.execute("SELECT CD_CBO, DS_OCUPACAO, ST_SITUACAO FROM DWCD_OCUPACAO")
    db_rows = cursor.fetchall()
    df_db = pd.DataFrame(db_rows, columns=["CD_CBO", "DS_OCUPACAO", "ST_SITUACAO"])

    # Data atual
    now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    # --- 1) Inserir novos (estão no df mas não no banco, exceto -1)
    novos = df_cbo[(df_cbo["cbo2002ocupacao"] != -1) &
                   (~df_cbo["cbo2002ocupacao"].isin(df_db["CD_CBO"]))]

    for _, row in novos.iterrows():
        cd_cbo = int(row["cbo2002ocupacao"])
        ds_ocup = row["cbo_descricao"]

        cursor.execute("""
            INSERT INTO DWCD_OCUPACAO (CD_CBO, DS_OCUPACAO, ST_SITUACAO, DT_CARGA)
            VALUES (?, ?, 'A', ?)
        """, (cd_cbo, ds_ocup, now))
        print(f"✅ Inserido ocupação {cd_cbo} - {ds_ocup}")

    # --- 2) Inativar (estão no banco mas não no df, exceto -1)
    inativos = df_db[(df_db["CD_CBO"] != -1) &
                     (~df_db["CD_CBO"].isin(df_cbo["cbo2002ocupacao"]))]

    for _, row in inativos.iterrows():
        cursor.execute("""
            UPDATE DWCD_OCUPACAO
            SET ST_SITUACAO = 'I', DT_CARGA = ?
            WHERE CD_CBO = ?
        """, (now, row["CD_CBO"]))
        print(f"⚠️ Inativada ocupação {row['CD_CBO']} - {row['DS_OCUPACAO']}")

    # --- 3) Atualizar descrição ou reativar (exceto -1)
    atualizacoes = df_cbo.merge(df_db, left_on="cbo2002ocupacao", right_on="CD_CBO")
    for _, row in atualizacoes.iterrows():
        if row["cbo2002ocupacao"] == -1:
            continue  # ignora o código especial -1

        cd_cbo = int(row["cbo2002ocupacao"])
        ds_ocup = row["cbo_descricao"]

        mudou_nome = (ds_ocup != row["DS_OCUPACAO"])
        estava_inativo = (row["ST_SITUACAO"] == "I")

        if mudou_nome or estava_inativo:
            cursor.execute("""
                UPDATE DWCD_OCUPACAO
                SET DS_OCUPACAO = ?, ST_SITUACAO = 'A', DT_CARGA = ?
                WHERE CD_CBO = ?
            """, (ds_ocup, now, cd_cbo))

            if estava_inativo:
                print(f"♻️ Reativada ocupação {cd_cbo} - {ds_ocup}")
            else:
                print(f"🔄 Atualizada ocupação {cd_cbo} - {ds_ocup}")

    conn.commit()
    conn.close()
    print("✅ Merge de ocupações finalizado com sucesso!")
```

```python
from datetime import datetime
import sqlite3, time

def carregar_movimento_incremental(df, db_path="Banco", periodo="ano"):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    inicio = time.time()
    total = len(df)
    print(f"🚀 Iniciando carga incremental de {total} registros...")

    # 1) Descobrir período da carga

    anos = df["DTOBITO"].dt.year.unique().tolist()
    condicoes = ",".join([f"'{a}'" for a in anos])
    delete_sql = f"DELETE FROM DWMV_OBITO WHERE strftime('%Y', DT_OBITO) IN ({condicoes})"

    # 2) Remover dados do período
    print(f"🗑️ Removendo registros do período selecionado ({periodo})...")
    cursor.execute(delete_sql)

    # 3) Inserir novos dados
    insert_sql = """
        INSERT INTO DWMV_OBITO (
            CD_OBITO, DT_OBITO, ST_ASSISTENCIA_MEDICA, ST_INVESTIGADO, ST_EPIDEMIA,
            DT_NASCIMENTO, IDADE, SK_CIRCUSNTANCIA, SK_MUNICIPIO_OBITO, SK_LOCAL_OBITO,
            SK_SEXO, SK_MUNICIPIO_RESIDENCIA, SK_RACA_COR, SK_OCUPACAO, SK_ESTADO_CIVIL,
            SK_ESCOLARIDADE, SK_TIPO_IDADE, DT_CARGA, SK_TERMINAL_1, SK_TERMINAL_2,
            SK_ANTECEDENTE_1, SK_ANTECEDENTE_2, SK_ANTECEDENTE_3, SK_ANTECEDENTE_4,
            SK_BASICA_1, SK_BASICA_2
        ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    """

    for i, row in df.iterrows():
        cursor.execute(insert_sql, (
            row["CD_OBITO"],
            row["DTOBITO"].strftime("%Y-%m-%d"),
            row.get("ASSISTMED", "I"),
            row.get("INVESTIG", "I"),
            row.get("STDOEPIDEM", "I"),
            row["DTNASC"].strftime("%Y-%m-%d"),
            row["NR_IDADE"],
            row.get("CIRCOBITO", -1),
            row["CODMUNOCOR"],
            row["LOCOCOR"],
            row["SEXO"],
            row["CODMUNRES"],
            row["RACACOR"],
            row["OCUP"],
            row["ESTCIV"],
            row["ESCFALAGR1"],
            row["TIPOIDADE"],
            datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            row["TERMINAL_1"], row["TERMINAL_2"],
            row["ANTECEDENTE_1"], row["ANTECEDENTE_2"], row["ANTECEDENTE_3"], row["ANTECEDENTE_4"],
            row["BASICA_1"], row["BASICA_2"]
        ))

        if (i+1) % 10000 == 0 or (i+1) == total:
            perc = ((i+1)/total)*100
            elapsed = time.time() - inicio
            print(f"🔄 {i+1}/{total} ({perc:.2f}%) processados - {elapsed:.1f}s")

    conn.commit()
    conn.close()
    fim = time.time()
    print(f"✅ Carga incremental concluída em {fim - inicio:.1f}s")
```
