# Tabela tipo causa

```python
# Conexão
conn = sqlite3.connect("Banco")
cursor = conn.cursor()

# Lista de causas a inserir
causas = [
    "Causa Terminal",
    "Causa Antecedente ou Consequencial",
    "Causa Básica"
]

for causa in causas:
    # Verifica se a causa já existe
    cursor.execute('''
        SELECT SK_TIPO_CAUSA FROM DWCD_TIPO_CAUSA
        WHERE DS_TIPO_CAUSA = ?
    ''', (causa,))
    existing = cursor.fetchone()

    if not existing:  # Só insere se não existir
        cursor.execute('''
            INSERT INTO DWCD_TIPO_CAUSA (DS_TIPO_CAUSA, ST_SITUACAO, DT_CARGA)
            VALUES (?, ?, ?)
        ''', (causa, 'A', datetime.now().strftime("%Y-%m-%d %H:%M:%S")))
        print(f"✅ Inserido: {causa}")
    else:
        print(f"⚠️ Já existe no banco: {causa}")

# Commit e fechar
conn.commit()
conn.close()
```

```python
def carregar_falecidos(df, db_path="Banco"):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Função auxiliar para normalizar ST
    def normalizar_st(valor):
        if valor == 1:
            return "S"
        elif valor == 2:
            return "N"
        elif valor in [9, -1]:
            return "I"
        return None

    # Pega SK próximo para falecido e óbito
    cursor.execute("SELECT MAX(SK_FALECIDO) FROM DWCD_FALECIDO")
    next_sk_falecido = (cursor.fetchone()[0] or 0) + 1

    cursor.execute("SELECT MAX(SK_OBITO) FROM DWMV_OBITO")
    next_sk_obito = (cursor.fetchone()[0] or 0) + 1

    for _, row in df.iterrows():
        # ======== IDADE ========
        idade_raw = str(row["IDADE"]).zfill(3)  # garante 3 dígitos
        cd_tipo_idade = int(idade_raw[0])       # 1º dígito
        ds_idade = int(idade_raw[1:])           # 2 últimos dígitos

        # Lookup SK_TIPO_IDADE
        cursor.execute("SELECT SK_TIPO_IDADE FROM DWCD_TIPO_IDADE WHERE CD_TIPO_IDADE = ?", (cd_tipo_idade,))
        sk_tipo_idade = cursor.fetchone()
        sk_tipo_idade = sk_tipo_idade[0] if sk_tipo_idade else None

        # Lookup Sexo
        cursor.execute("SELECT SK_SEXO FROM DWCD_SEXO WHERE CD_SEXO = ?", (row["SEXO"],))
        sk_sexo = cursor.fetchone()
        sk_sexo = sk_sexo[0] if sk_sexo else None

        # Lookup Raça/Cor
        cursor.execute("SELECT SK_RACA_COR FROM DWCD_RACA_COR WHERE CD_RACA_COR = ?", (row["RACACOR"],))
        sk_raca = cursor.fetchone()
        sk_raca = sk_raca[0] if sk_raca else None

        # Lookup Estado Civil
        cursor.execute("SELECT SK_ESTADO_CIVIL FROM DWCD_ESTADO_CIVIL WHERE CD_ESTADO_CIVIL = ?", (row["ESTCIV"],))
        sk_estciv = cursor.fetchone()
        sk_estciv = sk_estciv[0] if sk_estciv else None

        # Lookup Escolaridade
        cursor.execute("SELECT SK_ESCOLARIDADE FROM DWCD_ESCOLARIDADE WHERE CD_ESCOLARIDADE = ?", (row["ESCFALAGR1"],))
        sk_escolaridade = cursor.fetchone()
        sk_escolaridade = sk_escolaridade[0] if sk_escolaridade else None

        # Lookup Ocupação
        cursor.execute("SELECT SK_OCUPACAO FROM DWCD_OCUPACAO WHERE CD_CBO = ?", (row["OCUP"],))
        sk_ocup = cursor.fetchone()
        sk_ocup = sk_ocup[0] if sk_ocup else None

        # Lookup Municípios (nascimento e residência)
        cursor.execute("SELECT SK_MUNICIPIO FROM DWCD_MUNICIPIO WHERE CD_MUNICIPIO = ?", (row["CODMUNNATU"],))
        sk_mun_natu = cursor.fetchone()
        sk_mun_natu = sk_mun_natu[0] if sk_mun_natu else None

        cursor.execute("SELECT SK_MUNICIPIO FROM DWCD_MUNICIPIO WHERE CD_MUNICIPIO = ?", (row["CODMUNRES"],))
        sk_mun_res = cursor.fetchone()
        sk_mun_res = sk_mun_res[0] if sk_mun_res else None

        # ======== Inserir em DWCD_FALECIDO ========
        cursor.execute("""
            INSERT INTO DWCD_FALECIDO
            (SK_FALECIDO, SK_SEXO, SK_RACA_COR, SK_ESTADO_CIVIL, SK_ESCOLARIDADE,
             SK_OCUPACAO, SK_MUNICIPIO_NATURAL, SK_MUNICIPIO_RES, SK_TIPO_IDADE,
             DS_IDADE, DT_NASC, ST_SITUACAO, DT_CARGA)
            VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        """, (
            next_sk_falecido, sk_sexo, sk_raca, sk_estciv, sk_escolaridade,
            sk_ocup, sk_mun_natu, sk_mun_res, sk_tipo_idade,
            ds_idade, row["DTNASC"], 'A',
            datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        ))

        # ======== Inserir em DWMV_OBITO ========
        cursor.execute("""
            INSERT INTO DWMV_OBITO
            (SK_OBITO, SK_FALECIDO, DT_OBITO, HR_OBITO, SK_MUNICIPIO_OCOR,
             SK_LOCAL_OBITO, ST_ASSISTMED, ST_EPISINVEST, ST_EPIDEMIA,
             DT_ATESTADO, DT_CARGA)
            VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        """, (
            next_sk_obito, next_sk_falecido,
            row["DTOBITO"], row["HORAOBITO"], row["CODMUNOCOR"],
            row["LOCOCOR"], normalizar_st(row["ASSISTMED"]),
            normalizar_st(row.get("INVESTIG", -1)),
            normalizar_st(row.get("STDOEPIDEM", -1)),
            row["DTATESTADO"], datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        ))

        # ======== Inserir Causas ========
        causas = [
            ("Causa Básica", row["CAUSABAS"]),
            ("Causa Básica", row["CAUSABAS_O"]),
            ("Causa Terminal", row["LINHAA"]),
            ("Causa Antecedente ou Consequencial", row["LINHAB"]),
            ("Causa Antecedente ou Consequencial", row["LINHAC"]),
            ("Causa Antecedente ou Consequencial", row["LINHAD"]),
            ("Causa Antecedente ou Consequencial", row["LINHAII"]),
        ]
        for tipo, cid in causas:
            if cid not in [None, "-1", ""]:
                cid_norm = cid.replace("*", "").upper()

                cursor.execute("SELECT SK_TIPO_CAUSA FROM DWCD_TIPO_CAUSA WHERE DS_TIPO_CAUSA = ?", (tipo,))
                sk_tipo = cursor.fetchone()

                cursor.execute("SELECT SK_DOENÇA FROM DWCD_DOENÇA WHERE CD_CID = ?", (cid_norm,))
                sk_cid = cursor.fetchone()

                if sk_tipo and sk_cid:
                    cursor.execute("""
                        INSERT INTO DWMV_REL_OBITO_CAUSA
                        (SK_OBITO, SK_TIPO_CAUSA, SK_DOENÇA, DT_CARGA)
                        VALUES (?, ?, ?, ?)
                    """, (
                        next_sk_obito, sk_tipo[0], sk_cid[0],
                        datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                    ))

        # Incrementa os SKs para próxima linha
        next_sk_falecido += 1
        next_sk_obito += 1

    conn.commit()
    conn.close()
    print("✅ Carga concluída com sucesso!")
```
