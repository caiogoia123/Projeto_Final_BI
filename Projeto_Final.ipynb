# Criando banco

```python
conn = sqlite3.connect("Banco")
cursor = conn.cursor()

cursor.execute('''
  CREATE TABLE IF NOT EXISTS "DWCD_TIPO_CAUSA" (
    "SK_TIPO_CAUSA" INTEGER NOT NULL UNIQUE,
    "DS_TIPO_CAUSA" TEXT,
    "ST_SITUACAO" TEXT,
    "DT_CARGA" TEXT,
    PRIMARY KEY("SK_TIPO_CAUSA")
)''')

cursor.execute('''
  CREATE TABLE IF NOT EXISTS "DWCD_DOENÇA" (
    "SK_DOENÇA" INTEGER,
    "CD_CID" TEXT NOT NULL UNIQUE,
    "DS_DOENÇA" TEXT,
    "ST_SITUACAO" TEXT,
    "DT_CARGA" TEXT,
    PRIMARY KEY("SK_DOENÇA")
)''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS "DWCD_MUNICIPIO" (
    "SK_MUNICIPIO" INTEGER NOT NULL UNIQUE,
    "CD_MUNICIPIO" INTEGER,
    "NM_MUNICIPIO" TEXT,
    "SK_ESTADO" INTEGER,
    "DS_PORTE" TEXT,
    "ST_SITUACAO" TEXT,
    "DT_CARGA" TEXT,
    PRIMARY KEY("SK_MUNICIPIO")
)''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS "DWCD_ESTADO" (
    "SK_ESTADO" INTEGER NOT NULL UNIQUE,
    "CD_ESTADO" INTEGER,
    "CD_UF" TEXT,
    "DS_ESTADO" TEXT,
    "DS_REGIAO" INTEGER,
    "ST_SITUACAO" TEXT,
    "DT_CARGA" TEXT,
    PRIMARY KEY("SK_ESTADO")
)''')

cursor.execute('''
  CREATE TABLE IF NOT EXISTS "DWCD_OCUPACAO" (
    "SK_OCUPACAO" INTEGER NOT NULL UNIQUE,
    "CD_CBO" INTEGER,
    "DS_OCUPACAO" TEXT,
    "ST_SITUACAO" TEXT,
    "DT_CARGA" TEXT,
    PRIMARY KEY("SK_OCUPACAO")
)''')

cursor.execute('''
  CREATE TABLE IF NOT EXISTS "DWCD_OCUPACAO" (
    "SK_OCUPACAO" INTEGER NOT NULL UNIQUE,
    "CD_CBO" INTEGER,
    "DS_OCUPACAO" TEXT,
    "ST_SITUACAO" TEXT,
    "DT_CARGA" TEXT,
    PRIMARY KEY("SK_OCUPACAO")
)''')

cursor.execute('''
  CREATE TABLE IF NOT EXISTS "DWCD_SEXO" (
    "SK_SEXO" INTEGER NOT NULL UNIQUE,
    "CD_SEXO" INTEGER,
    "DS_SEXO" TEXT,
    "ST_SITUACAO" TEXT,
    "DT_CARGA" TEXT,
    PRIMARY KEY("SK_SEXO")
)''')

cursor.execute('''
  CREATE TABLE IF NOT EXISTS "DWCD_RACA_COR" (
    "SK_RACA_COR" INTEGER NOT NULL UNIQUE,
    "CD_RACA_COR" INTEGER,
    "DS_RACA_COR" TEXT,
    "ST_SITUACAO" TEXT,
    "DT_CARGA" TEXT,
    PRIMARY KEY("SK_RACA_COR")
)''')

cursor.execute('''
  CREATE TABLE IF NOT EXISTS "DWCD_ESTADO_CIVIL" (
    "SK_ESTADO_CIVIL" INTEGER NOT NULL UNIQUE,
    "CD_ESTADO_CIVIL" INTEGER,
    "DS_ESTADO_CIVIL" TEXT,
    "ST_SITUACAO" TEXT,
    "DT_CARGA" TEXT,
    PRIMARY KEY("SK_ESTADO_CIVIL")
)''')

cursor.execute('''
  CREATE TABLE IF NOT EXISTS "DWCD_ESCOLARIDADE" (
    "SK_ESCOLARIDADE" INTEGER NOT NULL UNIQUE,
    "CD_ESCOLARIDADE" INTEGER,
    "DS_ESCOLARIDADE" TEXT,
    "ST_SITUACAO" TEXT,
    "DT_CARGA" TEXT,
    PRIMARY KEY("SK_ESCOLARIDADE")
)''')

cursor.execute('''
  CREATE TABLE IF NOT EXISTS "DWCD_TIPO_IDADE" (
    "SK_TIPO_IDADE" INTEGER NOT NULL UNIQUE,
    "CD_TIPO_IDADE" INTEGER,
    "DS_TIPO_IDADE" TEXT,
    "ST_SITUACAO" TEXT,
    "DT_CARGA" TEXT,
    PRIMARY KEY("SK_TIPO_IDADE")
)''')

cursor.execute('''
  CREATE TABLE IF NOT EXISTS "DWCD_LOCAL_OBITO" (
    "SK_LOCAL_OBITO" INTEGER NOT NULL UNIQUE,
    "CD_LOCAL_OBITO" INTEGER,
    "DS_LOCAL_OBITO" TEXT,
    "ST_SITUACAO" TEXT,
    "DT_CARGA" TEXT,
    PRIMARY KEY("SK_LOCAL_OBITO")
)''')

# Tabela de dimensão do falecido
cursor.execute('''
CREATE TABLE IF NOT EXISTS "DWCD_FALECIDO" (
    "SK_FALECIDO" INTEGER NOT NULL UNIQUE,
    "SK_SEXO" INTEGER,
    "SK_RACA_COR" INTEGER,
    "SK_ESTADO_CIVIL" INTEGER,
    "SK_ESCOLARIDADE" INTEGER,
    "SK_OCUPACAO" INTEGER,
    "SK_MUNICIPIO_NATURAL" INTEGER,
    "SK_MUNICIPIO_RES" INTEGER,
    "SK_TIPO_IDADE" INTEGER,
    "DS_IDADE" INTEGER,
    "DT_NASC" TEXT,
    "ST_SITUACAO" TEXT,
    "DT_CARGA" TEXT,
    PRIMARY KEY("SK_FALECIDO")
)''')

# Tabela de fatos do óbito
cursor.execute('''
CREATE TABLE IF NOT EXISTS "DWMV_OBITO" (
    "SK_OBITO" INTEGER NOT NULL UNIQUE,
    "SK_FALECIDO" INTEGER,
    "DT_OBITO" TEXT,
    "HR_OBITO" TEXT,
    "SK_MUNICIPIO_OCOR" INTEGER,
    "SK_LOCAL_OBITO" INTEGER,
    "ST_ASSISTMED" TEXT,
    "ST_INVESTIGADO" TEXT,
    "ST_EPIDEMIA" TEXT,
    "DT_ATESTADO" TEXT,
    "DT_CARGA" TEXT,
    PRIMARY KEY("SK_OBITO")
)''')

# Tabela de relacionamento ÓBITO ↔ CAUSA
cursor.execute('''
CREATE TABLE IF NOT EXISTS "DWMV_REL_OBITO_CAUSA" (
    "SK_OBITO_CAUSA" INTEGER NOT NULL UNIQUE,
    "SK_OBITO" INTEGER,
    "SK_TIPO_CAUSA" INTEGER,
    "SK_DOENÇA" INTEGER,
    "DT_CARGA" TEXT,
    PRIMARY KEY("SK_OBITO_CAUSA")
)''')


conn.commit()
conn.close()
```

# Tabela tipo causa

```python
# Conexão
conn = sqlite3.connect("Banco")
cursor = conn.cursor()

# Lista de causas a inserir
causas = [
    "Causa Terminal",
    "Causa Antecedente ou Consequencial",
    "Causa Básica"
]

for causa in causas:
    # Verifica se a causa já existe
    cursor.execute('''
        SELECT SK_TIPO_CAUSA FROM DWCD_TIPO_CAUSA
        WHERE DS_TIPO_CAUSA = ?
    ''', (causa,))
    existing = cursor.fetchone()

    if not existing:  # Só insere se não existir
        cursor.execute('''
            INSERT INTO DWCD_TIPO_CAUSA (DS_TIPO_CAUSA, ST_SITUACAO, DT_CARGA)
            VALUES (?, ?, ?)
        ''', (causa, 'A', datetime.now().strftime("%Y-%m-%d %H:%M:%S")))
        print(f"✅ Inserido: {causa}")
    else:
        print(f"⚠️ Já existe no banco: {causa}")

# Commit e fechar
conn.commit()
conn.close()
```

# Tabela de município

```python
df_mun = pd.read_csv('lista_mun.csv', sep=';', encoding='latin-1')
```

```python
df_mun.tail()
```

```python
conn = sqlite3.connect("Banco")
cursor = conn.cursor()

for _, row in df_mun.iterrows():
    cd_mun = int(row["IBGE"]) if row["IBGE"] != -1 else -1
    nm_mun = row["Município"]
    porte = row["Porte"]

    # Extrair 2 primeiros dígitos do IBGE (CD_ESTADO)
    cd_estado = int(str(row["IBGE"]).zfill(7)[:2]) if row["IBGE"] != -1 else None

    # Buscar SK_ESTADO
    sk_estado = None
    if cd_estado:
        cursor.execute("SELECT SK_ESTADO FROM DWCD_ESTADO WHERE CD_ESTADO = ?", (cd_estado,))
        result = cursor.fetchone()
        sk_estado = result[0] if result else None

    cursor.execute("""
        INSERT OR IGNORE INTO DWCD_MUNICIPIO
        (CD_MUNICIPIO, NM_MUNICIPIO, SK_ESTADO, DS_PORTE, ST_SITUACAO, DT_CARGA)
        VALUES (?, ?, ?, ?, 'A', ?)
    """, (cd_mun, nm_mun, sk_estado, porte, datetime.now().strftime("%Y-%m-%d %H:%M:%S")))

conn.commit()
conn.close()
print("✅ Municípios inseridos com sucesso")
```

```python
# Conexão
conn = sqlite3.connect("Banco")
cursor = conn.cursor()

# Select para ver todos os registros
cursor.execute("SELECT * FROM DWCD_MUNICIPIO")
rows = cursor.fetchall()

# Nomes das colunas
colunas = [desc[0] for desc in cursor.description]

# Mostrar em formato de tabela (pandas)
df = pd.DataFrame(rows, columns=colunas)
print(df)

conn.close()
```

```python
conn = sqlite3.connect("Banco")
cursor = conn.cursor()

# Pega todos os municípios
cursor.execute("SELECT SK_MUNICIPIO, CD_MUNICIPIO FROM DWCD_MUNICIPIO")
municipios = cursor.fetchall()

print(municipios)

for sk_mun, cd_mun in municipios:
    cd_estado = str(cd_mun)[:2] # Convert to string before slicing
    print(cd_estado)
    # Busca o estado correspondente
    cursor.execute("SELECT SK_ESTADO FROM DWCD_ESTADO WHERE CD_ESTADO = ?", (cd_estado,))
    sk_estado = cursor.fetchone()[0]

    # Atualiza o SK_ESTADO
    cursor.execute("UPDATE DWCD_MUNICIPIO SET SK_ESTADO = ? WHERE SK_MUNICIPIO = ?", (sk_estado, sk_mun))

conn.commit()
conn.close()

print("✅ Municípios agora estão relacionados corretamente com seus estados")
```

```python
from datetime import datetime
import time

def carregar_falecidos(df, db_path="Banco"):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Função auxiliar para normalizar ST
    def normalizar_st(valor):
        if valor == 1:
            return "S"
        elif valor in [2, 0]:
            return "N"
        elif valor in [9, -1]:
            return "I"
        return None

    total = len(df)
    print(f"🚀 Iniciando carga de {total} registros...")

    inicio = time.time()

    # Pega SK próximo para falecido e óbito
    cursor.execute("SELECT MAX(SK_FALECIDO) FROM DWCD_FALECIDO")
    next_sk_falecido = (cursor.fetchone()[0] or 0) + 1

    cursor.execute("SELECT MAX(SK_OBITO) FROM DWMV_OBITO")
    next_sk_obito = (cursor.fetchone()[0] or 0) + 1

    for i, row in df.iterrows():

        # Lookup SK_TIPO_IDADE
        cursor.execute("SELECT SK_TIPO_IDADE FROM DWCD_TIPO_IDADE WHERE CD_TIPO_IDADE = ?", (row["TIPOIDADE"],))
        sk_tipo_idade = cursor.fetchone()
        sk_tipo_idade = sk_tipo_idade[0] if sk_tipo_idade else None

        # Lookup Sexo
        cursor.execute("SELECT SK_SEXO FROM DWCD_SEXO WHERE CD_SEXO = ?", (row["SEXO"],))
        sk_sexo = cursor.fetchone()
        sk_sexo = sk_sexo[0] if sk_sexo else None

        # Lookup Raça/Cor
        cursor.execute("SELECT SK_RACA_COR FROM DWCD_RACA_COR WHERE CD_RACA_COR = ?", (row["RACACOR"],))
        sk_raca = cursor.fetchone()
        sk_raca = sk_raca[0] if sk_raca else None

        # Lookup Estado Civil
        cursor.execute("SELECT SK_ESTADO_CIVIL FROM DWCD_ESTADO_CIVIL WHERE CD_ESTADO_CIVIL = ?", (row["ESTCIV"],))
        sk_estciv = cursor.fetchone()
        sk_estciv = sk_estciv[0] if sk_estciv else None

        # Lookup Escolaridade
        cursor.execute("SELECT SK_ESCOLARIDADE FROM DWCD_ESCOLARIDADE WHERE CD_ESCOLARIDADE = ?", (row["ESCFALAGR1"],))
        sk_escolaridade = cursor.fetchone()
        sk_escolaridade = sk_escolaridade[0] if sk_escolaridade else None

        # Lookup Ocupação
        cursor.execute("SELECT SK_OCUPACAO FROM DWCD_OCUPACAO WHERE CD_CBO = ?", (row["OCUP"],))
        sk_ocup = cursor.fetchone()
        sk_ocup = sk_ocup[0] if sk_ocup else None

        # Lookup Municípios (nascimento e residência)
        cursor.execute("SELECT SK_MUNICIPIO FROM DWCD_MUNICIPIO WHERE CD_MUNICIPIO = ?", (row["CODMUNNATU"],))
        sk_mun_natu = cursor.fetchone()
        sk_mun_natu = sk_mun_natu[0] if sk_mun_natu else None

        cursor.execute("SELECT SK_MUNICIPIO FROM DWCD_MUNICIPIO WHERE CD_MUNICIPIO = ?", (row["CODMUNRES"],))
        sk_mun_res = cursor.fetchone()
        sk_mun_res = sk_mun_res[0] if sk_mun_res else None

        # Lookup Município Ocorrência
        cursor.execute("SELECT SK_MUNICIPIO FROM DWCD_MUNICIPIO WHERE CD_MUNICIPIO = ?", (row["CODMUNOCOR"],))
        sk_mun_ocor = cursor.fetchone()
        sk_mun_ocor = sk_mun_ocor[0] if sk_mun_ocor else None

        # ======== Inserir em DWCD_FALECIDO ========
        cursor.execute("""
            INSERT INTO DWCD_FALECIDO
            (SK_FALECIDO, SK_SEXO, SK_RACA_COR, SK_ESTADO_CIVIL, SK_ESCOLARIDADE,
             SK_OCUPACAO, SK_MUNICIPIO_NATURAL, SK_MUNICIPIO_RES, SK_TIPO_IDADE,
             DS_IDADE, DT_NASC, ST_SITUACAO, DT_CARGA)
            VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        """, (
            next_sk_falecido, sk_sexo, sk_raca, sk_estciv, sk_escolaridade,
            sk_ocup, sk_mun_natu, sk_mun_res, sk_tipo_idade,
            row["NR_IDADE"], row["DTNASC"], 'A',
            datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        ))

        # ======== Inserir em DWMV_OBITO ========
        cursor.execute("""
            INSERT INTO DWMV_OBITO
            (SK_OBITO, SK_FALECIDO, DT_OBITO, HR_OBITO, SK_MUNICIPIO_OCOR,
             SK_LOCAL_OBITO, ST_ASSISTMED, ST_INVESTIGADO, ST_EPIDEMIA,
             DT_ATESTADO, DT_CARGA)
            VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        """, (
            next_sk_obito, next_sk_falecido,
            row["DTOBITO"], row["HORAOBITO"], sk_mun_ocor, # Corrected: Use sk_mun_ocor
            row["LOCOCOR"], normalizar_st(row["ASSISTMED"]),
            normalizar_st(row.get("INVESTIG", -1)),
            normalizar_st(row.get("STDOEPIDEM", -1)),
            row["DTATESTADO"], datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        ))

        # ======== Inserir Causas ========
        causas = [
            ("Causa Básica", row["CAUSABAS"]),
            ("Causa Básica", row["CAUSABAS_O"]),
            ("Causa Terminal", row["LINHAA"]),
            ("Causa Antecedente ou Consequencial", row["LINHAB"]),
            ("Causa Antecedente ou Consequencial", row["LINHAC"]),
            ("Causa Antecedente ou Consequencial", row["LINHAD"]),
            ("Causa Antecedente ou Consequencial", row["LINHAII"]),
        ]
        for tipo, cid in causas:
            if cid not in [None, "-1", ""]:
                cid_norm = cid.replace("*", "").upper()

                cursor.execute("SELECT SK_TIPO_CAUSA FROM DWCD_TIPO_CAUSA WHERE DS_TIPO_CAUSA = ?", (tipo,))
                sk_tipo = cursor.fetchone()

                cursor.execute("SELECT SK_DOENÇA FROM DWCD_DOENÇA WHERE CD_CID = ?", (cid_norm,))
                sk_cid = cursor.fetchone()

                if sk_tipo and sk_cid:
                    cursor.execute("""
                        INSERT INTO DWMV_REL_OBITO_CAUSA
                        (SK_OBITO, SK_TIPO_CAUSA, SK_DOENÇA, DT_CARGA)
                        VALUES (?, ?, ?, ?)
                    """, (
                        next_sk_obito, sk_tipo[0], sk_cid[0],
                        datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                    ))

        # Incrementa os SKs para próxima linha
        next_sk_falecido += 1
        next_sk_obito += 1

        # ======== Prints de progresso ========
        if (i + 1) % 1000 == 0 or (i + 1) == total:
            perc = ((i + 1) / total) * 100
            elapsed = time.time() - inicio
            print(f"🔄 Processados {i+1}/{total} ({perc:.2f}%) - Tempo decorrido: {elapsed:.1f}s")

    conn.commit()
    conn.close()
    fim = time.time()
    print(f"✅ Carga concluída! Tempo total: {fim - inicio:.1f}s")
```
