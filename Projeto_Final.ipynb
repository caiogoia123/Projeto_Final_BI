# Tabela tipo causa

```python
# Conexão
conn = sqlite3.connect("Banco")
cursor = conn.cursor()

# Lista de causas a inserir
causas = [
    "Causa Terminal",
    "Causa Antecedente ou Consequencial",
    "Causa Básica"
]

for causa in causas:
    # Verifica se a causa já existe
    cursor.execute('''
        SELECT SK_TIPO_CAUSA FROM DWCD_TIPO_CAUSA
        WHERE DS_TIPO_CAUSA = ?
    ''', (causa,))
    existing = cursor.fetchone()

    if not existing:  # Só insere se não existir
        cursor.execute('''
            INSERT INTO DWCD_TIPO_CAUSA (DS_TIPO_CAUSA, ST_SITUACAO, DT_CARGA)
            VALUES (?, ?, ?)
        ''', (causa, 'A', datetime.now().strftime("%Y-%m-%d %H:%M:%S")))
        print(f"✅ Inserido: {causa}")
    else:
        print(f"⚠️ Já existe no banco: {causa}")

# Commit e fechar
conn.commit()
conn.close()
```

```python
# Conexão
conn = sqlite3.connect("Banco")
cursor = conn.cursor()

# Select para ver todos os registros
cursor.execute("SELECT * FROM DWCD_TIPO_CAUSA")
rows = cursor.fetchall()

# Nomes das colunas
colunas = [desc[0] for desc in cursor.description]

# Mostrar em formato de tabela (pandas)
df = pd.DataFrame(rows, columns=colunas)
print(df)

conn.close()
```

# Tabela CID

```python
# Carregar o arquivo JSON
with open("api.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Transformar em DataFrame
df_cid = pd.DataFrame(data["data"])
```

```python
colunas_excluir = [
    "codigo_nome"

]

df_cid = df_cid.drop(columns=colunas_excluir, errors="ignore")
```

```python
# Remove todos os pontos da coluna 'codigo'
df_cid["codigo"] = df_cid["codigo"].str.replace(".", "", regex=False)

# Conferir resultado
print(df_cid.head())
```

```python
conn = sqlite3.connect("Banco")
cursor = conn.cursor()

insert_query = """
INSERT OR IGNORE INTO DWCD_DOENÇA (CD_CID, DS_DOENÇA, ST_SITUACAO, DT_CARGA)
VALUES (?, ?, 'A', ?)
"""

data = df_cid[['codigo','nome']].values.tolist()
current_datetime = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

# Add the current datetime to each row in the data list
data_with_datetime = [row + [current_datetime] for row in data]

cursor.executemany(insert_query, data_with_datetime)
conn.commit()
conn.close()
```

```python
# Conexão
conn = sqlite3.connect("Banco")
cursor = conn.cursor()

# Select para ver todos os registros
cursor.execute("SELECT * FROM DWCD_DOENÇA")
rows = cursor.fetchall()

# Nomes das colunas
colunas = [desc[0] for desc in cursor.description]

# Mostrar em formato de tabela (pandas)
df = pd.DataFrame(rows, columns=colunas)

conn.close()
```
