```python
def merge_cid(df_cid, db_path="Banco"):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Pega dados atuais do banco
    cursor.execute("SELECT CD_CID, DS_DOENcA, ST_SITUACAO FROM DWCD_DOENcA")
    db_rows = cursor.fetchall()
    df_db = pd.DataFrame(db_rows, columns=["CD_CID", "DS_DOENcA", "ST_SITUACAO"])

    # Garantir que c√≥digo esteja em string sem pontos
    df_cid["codigo"] = df_cid["codigo"].astype(str).str.replace(".", "", regex=False)

    # Data atual
    now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    # --- 1) Inserir novos (est√£o na origem mas n√£o no banco)
    novos = df_cid[(df_cid['codigo'] != -1) &
                  (~df_cid["codigo"].isin(df_db["CD_CID"]))]
    for _, row in novos.iterrows():
        cursor.execute("""
            INSERT INTO DWCD_DOENcA (CD_CID, DS_DOENcA, ST_SITUACAO, DT_CARGA)
            VALUES (?, ?, 'A', ?)
        """, (row["codigo"], row["nome"], now))
        print(f"‚úÖ Inserido novo CID {row['codigo']}")

    # --- 2) Inativar (est√£o no banco mas n√£o na origem)
    inativos = df_db[(df_db["CD_CID"] != -1) &
                    (~df_db["CD_CID"].isin(df_cid["codigo"]))]
    for _, row in inativos.iterrows():
        cursor.execute("""
            UPDATE DWCD_DOENcA
            SET ST_SITUACAO = 'I', DT_CARGA = ?
            WHERE CD_CID = ?
        """, (now, row["CD_CID"]))
        print(f"‚ö†Ô∏è Inativado CID {row['CD_CID']}")

    # --- 3) Atualizar descric√£o ou reativar se necess√°rio
    atualizacoes = df_cid.merge(df_db, left_on="codigo", right_on="CD_CID")
    for _, row in atualizacoes.iterrows():
        # Caso a descri√ß√£o tenha mudado ou esteja inativa mas voltou ao df
        if (row["nome"] != row["DS_DOEN√áA"]) or (row["ST_SITUACAO"] == "I"):
            cursor.execute("""
                UPDATE DWCD_DOENcA
                SET DS_DOENcA = ?, ST_SITUACAO = 'A', DT_CARGA = ?
                WHERE CD_CID = ?
            """, (row["nome"], now, row["codigo"]))

            if row["ST_SITUACAO"] == "I":
                print(f"‚ôªÔ∏è Reativado CID {row['codigo']}")
            else:
                print(f"üîÑ Atualizado CID {row['codigo']}")

    conn.commit()
    conn.close()
    print("‚úÖ Merge finalizado com sucesso!")
```

```python
def merge_municipios(df_mun, db_path="Banco"):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Pega dados atuais do banco
    cursor.execute("SELECT CD_MUNICIPIO, CD_MUNICIPIO, DS_PORTE, SK_ESTADO, ST_SITUACAO FROM DWCD_MUNICIPIO")
    db_rows = cursor.fetchall()
    df_db = pd.DataFrame(db_rows, columns=["CD_MUNICIPIO", "CD_MUNICIPIO", "DS_PORTE", "SK_ESTADO", "ST_SITUACAO"])

    # Data atual
    now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    # --- 1) Inserir novos (ignorando -1)
    novos = df_mun[(df_mun["IBGE"] != -1) & (~df_mun["IBGE"].isin(df_db["CD_MUNICIPIO"]))]
    for _, row in novos.iterrows():
        cd_mun = int(row["IBGE"])
        nm_mun = row["Munic√≠pio"]
        porte = row["Porte"]

        # Descobre SK_ESTADO pelo c√≥digo IBGE
        cd_estado = int(str(row["IBGE"]).zfill(7)[:2])
        cursor.execute("SELECT SK_ESTADO FROM DWCD_ESTADO WHERE CD_ESTADO = ?", (cd_estado,))
        res = cursor.fetchone()
        sk_estado = res[0] if res else -1

        cursor.execute("""
            INSERT INTO DWCD_MUNICIPIO (CD_MUNICIPIO, CD_MUNICIPIO, SK_ESTADO, DS_PORTE, ST_SITUACAO, DT_CARGA)
            VALUES (?, ?, ?, ?, 'A', ?)
        """, (cd_mun, nm_mun, sk_estado, porte, now))
        print(f"‚úÖ Inserido munic√≠pio {cd_mun} - {nm_mun}")

    # --- 2) Inativar (existem no banco mas n√£o no df_mun, ignorando -1)
    inativos = df_db[(df_db["CD_MUNICIPIO"] != -1) & (~df_db["CD_MUNICIPIO"].isin(df_mun["IBGE"]))]
    for _, row in inativos.iterrows():
        cursor.execute("""
            UPDATE DWCD_MUNICIPIO
            SET ST_SITUACAO = 'I', DT_CARGA = ?
            WHERE CD_MUNICIPIO = ?
        """, (now, row["CD_MUNICIPIO"]))
        print(f"‚ö†Ô∏è Inativado munic√≠pio {row['CD_MUNICIPIO']}")

    # --- 3) Atualizar descri√ß√£o/porte/estado ou reativar (ignorando -1)
    atualizacoes = df_mun.merge(df_db, left_on="IBGE", right_on="CD_MUNICIPIO")
    for _, row in atualizacoes.iterrows():
        if row["IBGE"] == -1:
            continue  # ignora "N√£o informado"

        cd_mun = int(row["IBGE"])
        nm_mun = row["Munic√≠pio"]
        porte = row["Porte"]

        # Descobre SK_ESTADO atualizado
        cd_estado = int(str(row["IBGE"]).zfill(7)[:2])
        cursor.execute("SELECT SK_ESTADO FROM DWCD_ESTADO WHERE CD_ESTADO = ?", (cd_estado,))
        res = cursor.fetchone()
        sk_estado = res[0] if res else -1

        mudou_nome = (nm_mun != row["CD_MUNICIPIO"])
        mudou_porte = (porte != row["DS_PORTE"])
        mudou_estado = (sk_estado != row["SK_ESTADO"])
        estava_inativo = (row["ST_SITUACAO"] == "I")

        if mudou_nome or mudou_porte or mudou_estado or estava_inativo:
            cursor.execute("""
                UPDATE DWCD_MUNICIPIO
                SET CD_MUNICIPIO = ?, DS_PORTE = ?, SK_ESTADO = ?, ST_SITUACAO = 'A', DT_CARGA = ?
                WHERE CD_MUNICIPIO = ?
            """, (nm_mun, porte, sk_estado, now, cd_mun))

            if estava_inativo:
                print(f"‚ôªÔ∏è Reativado munic√≠pio {cd_mun} - {nm_mun}")
            else:
                print(f"üîÑ Atualizado munic√≠pio {cd_mun} - {nm_mun}")

    conn.commit()
    conn.close()
    print("‚úÖ Merge de munic√≠pios finalizado com sucesso!")
```

```python
def merge_ocupacao(df_cbo, db_path="Banco"):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Pega dados atuais do banco
    cursor.execute("SELECT CD_CBO, DS_OCUPACAO, ST_SITUACAO FROM DWCD_OCUPACAO")
    db_rows = cursor.fetchall()
    df_db = pd.DataFrame(db_rows, columns=["CD_CBO", "DS_OCUPACAO", "ST_SITUACAO"])

    # Data atual
    now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    # --- 1) Inserir novos (est√£o no df mas n√£o no banco, exceto -1)
    novos = df_cbo[(df_cbo["cbo2002ocupacao"] != -1) &
                   (~df_cbo["cbo2002ocupacao"].isin(df_db["CD_CBO"]))]

    for _, row in novos.iterrows():
        cd_cbo = int(row["cbo2002ocupacao"])
        ds_ocup = row["cbo_descricao"]

        cursor.execute("""
            INSERT INTO DWCD_OCUPACAO (CD_CBO, DS_OCUPACAO, ST_SITUACAO, DT_CARGA)
            VALUES (?, ?, 'A', ?)
        """, (cd_cbo, ds_ocup, now))
        print(f"‚úÖ Inserido ocupa√ß√£o {cd_cbo} - {ds_ocup}")

    # --- 2) Inativar (est√£o no banco mas n√£o no df, exceto -1)
    inativos = df_db[(df_db["CD_CBO"] != -1) &
                     (~df_db["CD_CBO"].isin(df_cbo["cbo2002ocupacao"]))]

    for _, row in inativos.iterrows():
        cursor.execute("""
            UPDATE DWCD_OCUPACAO
            SET ST_SITUACAO = 'I', DT_CARGA = ?
            WHERE CD_CBO = ?
        """, (now, row["CD_CBO"]))
        print(f"‚ö†Ô∏è Inativada ocupa√ß√£o {row['CD_CBO']} - {row['DS_OCUPACAO']}")

    # --- 3) Atualizar descri√ß√£o ou reativar (exceto -1)
    atualizacoes = df_cbo.merge(df_db, left_on="cbo2002ocupacao", right_on="CD_CBO")
    for _, row in atualizacoes.iterrows():
        if row["cbo2002ocupacao"] == -1:
            continue  # ignora o c√≥digo especial -1

        cd_cbo = int(row["cbo2002ocupacao"])
        ds_ocup = row["cbo_descricao"]

        mudou_nome = (ds_ocup != row["DS_OCUPACAO"])
        estava_inativo = (row["ST_SITUACAO"] == "I")

        if mudou_nome or estava_inativo:
            cursor.execute("""
                UPDATE DWCD_OCUPACAO
                SET DS_OCUPACAO = ?, ST_SITUACAO = 'A', DT_CARGA = ?
                WHERE CD_CBO = ?
            """, (ds_ocup, now, cd_cbo))

            if estava_inativo:
                print(f"‚ôªÔ∏è Reativada ocupa√ß√£o {cd_cbo} - {ds_ocup}")
            else:
                print(f"üîÑ Atualizada ocupa√ß√£o {cd_cbo} - {ds_ocup}")

    conn.commit()
    conn.close()
    print("‚úÖ Merge de ocupa√ß√µes finalizado com sucesso!")
```

```python
from datetime import datetime
import sqlite3, time
import pandas as pd

def carregar_movimento_incremental(df, db_path="Banco", periodo="ano"):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    inicio = time.time()
    total = len(df)
    print(f"üöÄ Iniciando carga incremental de {total} registros...")

    # Convert date columns to datetime
    df['DTOBITO'] = pd.to_datetime(df['DTOBITO'], errors='coerce')
    df['DTNASC'] = pd.to_datetime(df['DTNASC'], errors='coerce')

    # === Fun√ß√£o auxiliar para normalizar colunas ST ===
    def normalizar_st(valor):
        if valor == 1:
            return "S"
        elif valor in [0,2]:
            return "N"
        elif valor in [9, -1]:
            return "I"
        return None  # caso n√£o esteja nos valores esperados

    # 1) Descobrir per√≠odo da carga
    anos = pd.to_datetime(df["DTOBITO"]).dt.year.unique().tolist()
    condicoes = ",".join([f"'{a}'" for a in anos])
    delete_sql = f"DELETE FROM DWMV_OBITO WHERE strftime('%Y', DT_OBITO) IN ({condicoes})"

    # 2) Remover dados do per√≠odo
    print(f"üóëÔ∏è Removendo registros do per√≠odo selecionado ({anos})...")
    cursor.execute(delete_sql)

    # 3) Inserir novos dados
    insert_sql = """
        INSERT INTO DWMV_OBITO (
            CD_OBITO, DT_OBITO, ST_ASSISTENCIA_MEDICA, ST_INVESTIGADO, ST_EPIDEMIA,
            DT_NASCIMENTO, IDADE, SK_CIRCUSNTANCIA, SK_MUNICIPIO_OBITO, SK_LOCAL_OBITO,
            SK_SEXO, SK_MUNICIPIO_RESIDENCIA, SK_RACA_COR, SK_OCUPACAO, SK_ESTADO_CIVIL,
            SK_ESCOLARIDADE, SK_TIPO_IDADE, DT_CARGA, SK_TERMINAL_1, SK_TERMINAL_2,
            SK_ANTECEDENTE_1, SK_ANTECEDENTE_2, SK_ANTECEDENTE_3, SK_ANTECEDENTE_4,
            SK_BASICA_1, SK_BASICA_2
        ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    """

    def lookup(table, cd_field, cd_value, sk_field):
        if cd_value in [None, "", -1, " ", "nan"]:
            return -1
        cursor.execute(f"SELECT {sk_field} FROM {table} WHERE {cd_field} = ?", (cd_value,))
        res = cursor.fetchone()
        return res[0] if res else -1

    for i, row in df.iterrows():

        # === Lookups com fallback -1 ===
        sk_circ = lookup("DWCD_CIRCUNSTANCIA", "CD_CIRCUNSTANCIA", row.get("CIRCOBITO", -1), "SK_CIRCUSTANCIA")
        sk_mun_obito = lookup("DWCD_MUNICIPIO", "CD_MUNICIPIO", row["CODMUNOCOR"], "SK_MUNICIPIO")
        sk_mun_res = lookup("DWCD_MUNICIPIO", "CD_MUNICIPIO", row["CODMUNRES"], "SK_MUNICIPIO")
        sk_local = lookup("DWCD_LOCAL_OBITO", "CD_LOCAL_OBITO", row["LOCOCOR"], "SK_LOCAL_OBITO")
        sk_sexo = lookup("DWCD_SEXO", "CD_SEXO", row["SEXO"], "SK_SEXO")
        sk_raca = lookup("DWCD_RACA_COR", "CD_RACA_COR", row["RACACOR"], "SK_RACA_COR")
        sk_ocup = lookup("DWCD_OCUPACAO", "CD_CBO", row["OCUP"], "SK_OCUPACAO")
        sk_estciv = lookup("DWCD_ESTADO_CIVIL", "CD_ESTADO_CIVIL", row["ESTCIV"], "SK_ESTADO_CIVIL")
        sk_escolar = lookup("DWCD_ESCOLARIDADE", "CD_ESCOLARIDADE", row["ESCFALAGR1"], "SK_ESCOLARIDADE")
        sk_tipoidade = lookup("DWCD_TIPO_IDADE", "CD_TIPO_IDADE", row["TIPOIDADE"], "SK_TIPO_IDADE")

        # === Lookups das causas (CID ‚Üí SK_DOENCA) ===
        sk_terminal_1 = lookup("DWCD_DOENCA", "CD_CID", row["TERMINAL_1"], "SK_DOENCA")
        sk_terminal_2 = lookup("DWCD_DOENCA", "CD_CID", row["TERMINAL_2"], "SK_DOENCA")
        sk_antec_1 = lookup("DWCD_DOENCA", "CD_CID", row["ANTECEDENTE_1"], "SK_DOENCA")
        sk_antec_2 = lookup("DWCD_DOENCA", "CD_CID", row["ANTECEDENTE_2"], "SK_DOENCA")
        sk_antec_3 = lookup("DWCD_DOENCA", "CD_CID", row["ANTECEDENTE_3"], "SK_DOENCA")
        sk_antec_4 = lookup("DWCD_DOENCA", "CD_CID", row["ANTECEDENTE_4"], "SK_DOENCA")
        sk_basica_1 = lookup("DWCD_DOENCA", "CD_CID", row["BASICA_1"], "SK_DOENCA")
        sk_basica_2 = lookup("DWCD_DOENCA", "CD_CID", row["BASICA_2"], "SK_DOENCA")


        # === Normalizar colunas ST antes da inser√ß√£o ===
        st_assist = normalizar_st(row.get("ASSISTMED", -1))
        st_invest = normalizar_st(row.get("INVESTIG", -1))
        st_epidem = normalizar_st(row.get("STDOEPIDEM", -1))


        # === Inser√ß√£o final ===
        cursor.execute(insert_sql, (
            row["CD_OBITO"],
            row["DTOBITO"].strftime("%Y-%m-%d"),
            st_assist,
            st_invest,
            st_epidem,
            row["DTNASC"].strftime("%Y-%m-%d"),
            row["NR_IDADE"],
            sk_circ,
            sk_mun_obito,
            sk_local,
            sk_sexo,
            sk_mun_res,
            sk_raca,
            sk_ocup,
            sk_estciv,
            sk_escolar,
            sk_tipoidade,
            datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            sk_terminal_1, sk_terminal_2,
            sk_antec_1, sk_antec_2, sk_antec_3, sk_antec_4,
            sk_basica_1, sk_basica_2
        ))

        if (i+1) % 10000 == 0 or (i+1) == total:
            perc = ((i+1)/total)*100
            elapsed = time.time() - inicio
            print(f"üîÑ {i+1}/{total} ({perc:.2f}%) processados - {elapsed:.1f}s")

    conn.commit()
    conn.close()
    fim = time.time()
    print(f"‚úÖ Carga incremental conclu√≠da em {fim - inicio:.1f}s")
```

```python
conn = sqlite3.connect("Banco")

query = """
SELECT
    ob.CD_OBITO,
    ob.DT_OBITO,
    ob.DT_NASCIMENTO,
    ob.IDADE,
    ob.ST_ASSISTENCIA_MEDICA,
    ob.ST_INVESTIGADO,
    ob.ST_EPIDEMIA,

    -- Circunst√¢ncia
    circ.DS_CIRCUNSTANCIA,

    -- Sexo
    s.DS_SEXO,

    -- RaCa/Cor
    r.DS_RACA_COR,

    -- Estado civil
    ec.DS_ESTADO_CIVIL,

    -- Escolaridade
    esc.DS_ESCOLARIDADE,

    -- OcupaC√£o
    o.DS_OCUPACAO,

    -- Tipo de idade
    ti.DS_TIPO_IDADE,

    -- Munic√≠pios
    mo.DS_MUNICIPIO AS MUNICIPIO_OBITO,
    mr.DS_MUNICIPIO AS MUNICIPIO_RESIDENCIA,

    -- Local do √≥bito
    lo.DS_LOCAL_OBITO,

    -- Causas (joins com doenCas)
    dt1.CD_CID AS CID_TERMINAL_1, dt1.DS_DOENCA AS DOENCA_TERMINAL_1,
    dt2.CD_CID AS CID_TERMINAL_2, dt2.DS_DOENCA AS DOENCA_TERMINAL_2,

    da1.CD_CID AS CID_ANTECEDENTE_1, da1.DS_DOENCA AS DOENCA_ANTECEDENTE_1,
    da2.CD_CID AS CID_ANTECEDENTE_2, da2.DS_DOENCA AS DOENCA_ANTECEDENTE_2,
    da3.CD_CID AS CID_ANTECEDENTE_3, da3.DS_DOENCA AS DOENCA_ANTECEDENTE_3,
    da4.CD_CID AS CID_ANTECEDENTE_4, da4.DS_DOENCA AS DOENCA_ANTECEDENTE_4,

    db1.CD_CID AS CID_BASICA_1, db1.DS_DOENCA AS DOENCA_BASICA_1,
    db2.CD_CID AS CID_BASICA_2, db2.DS_DOENCA AS DOENCA_BASICA_2,

    ob.DT_CARGA

FROM DWMV_OBITO ob
LEFT JOIN DWCD_CIRCUNSTANCIA circ ON ob.SK_CIRCUSNTANCIA = circ.SK_CIRCUSTANCIA
LEFT JOIN DWCD_SEXO s              ON ob.SK_SEXO = s.SK_SEXO
LEFT JOIN DWCD_RACA_COR r          ON ob.SK_RACA_COR = r.SK_RACA_COR
LEFT JOIN DWCD_ESTADO_CIVIL ec     ON ob.SK_ESTADO_CIVIL = ec.SK_ESTADO_CIVIL
LEFT JOIN DWCD_ESCOLARIDADE esc    ON ob.SK_ESCOLARIDADE = esc.SK_ESCOLARIDADE
LEFT JOIN DWCD_OCUPACAO o          ON ob.SK_OCUPACAO = o.SK_OCUPACAO
LEFT JOIN DWCD_TIPO_IDADE ti       ON ob.SK_TIPO_IDADE = ti.SK_TIPO_IDADE
LEFT JOIN DWCD_MUNICIPIO mo        ON ob.SK_MUNICIPIO_OBITO = mo.SK_MUNICIPIO
LEFT JOIN DWCD_MUNICIPIO mr        ON ob.SK_MUNICIPIO_RESIDENCIA = mr.SK_MUNICIPIO
LEFT JOIN DWCD_LOCAL_OBITO lo      ON ob.SK_LOCAL_OBITO = lo.SK_LOCAL_OBITO

-- Joins para doenCas (CID)
LEFT JOIN DWCD_DOENCA dt1 ON ob.SK_TERMINAL_1     = dt1.SK_DOENCA
LEFT JOIN DWCD_DOENCA dt2 ON ob.SK_TERMINAL_2     = dt2.SK_DOENCA

LEFT JOIN DWCD_DOENCA da1 ON ob.SK_ANTECEDENTE_1  = da1.SK_DOENCA
LEFT JOIN DWCD_DOENCA da2 ON ob.SK_ANTECEDENTE_2  = da2.SK_DOENCA
LEFT JOIN DWCD_DOENCA da3 ON ob.SK_ANTECEDENTE_3  = da3.SK_DOENCA
LEFT JOIN DWCD_DOENCA da4 ON ob.SK_ANTECEDENTE_4  = da4.SK_DOENCA

LEFT JOIN DWCD_DOENCA db1 ON ob.SK_BASICA_1       = db1.SK_DOENCA
LEFT JOIN DWCD_DOENCA db2 ON ob.SK_BASICA_2       = db2.SK_DOENCA

WHERE ob.CD_OBITO = 		32132022;  -- exemplo
"""

df_valida = pd.read_sql_query(query, conn)
conn.close()

df_valida.head().T
```
